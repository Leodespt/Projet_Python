```python
# import pandas as pd   
# import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
# import os

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
url = 'https://github.com/Leodespt/Projet_Python/blob/main/ObesityDataSet_raw_and_data_sinthetic.csv?raw=true'
df = pd.read_csv(url)
df
```

```python
count_by_SMOKE=df.groupby(by="SMOKE").size()
count_by_SMOKE


count_by_Age=df.groupby(by="Age").size()
count_by_Age
```

## LOCAL_ANALYSIS


### LOCAL_ANALYSIS/Harmful_addictions

```python

smoker = len(df.loc[df['SMOKE'] == 'yes'])
nonsmoker = len(df.loc[df['SMOKE'] == 'no'])

CALC_no = len(df.loc[df['CALC'] == 'no'])
CALC_Sometimes = len(df.loc[df['CALC'] == 'Sometimes'])
CALC_Frequently = len(df.loc[df['CALC'] == 'Frequently'])
CALC_Always = len(df.loc[df['CALC'] == 'Always'])

fig, ax = plt.subplots(1,2, figsize=(15,10))

ax[0].pie([smoker,nonsmoker],labels=['Smoker','Non Smoker'],
    autopct='%1.1f%%', startangle = -60,explode=[0.2,0], shadow=True)
ax[0].set_title('Proportion of Smoker')

ax[1].bar(['no','Sometimes','Frequently','Always'],
    [CALC_no,CALC_Sometimes,CALC_Frequently,CALC_Always], 
    color = 'cyan')
ax[1].set_title('Consumption of alcohol\n')

ax[0].legend(loc=3)

plt.tight_layout()
```

### LOCAL_ANALYSIS/Good_eating_habits

```python
df_FCVC = df[["FCVC"]]

df_CH2O = df[["CH2O"]]

fig, ax = plt.subplots(1,2, figsize=(15,10))

df_CH2O.boxplot(showmeans=True,ax=ax[1],grid=False)
ax[1].set_title("Consumption of water daily\n")

df_FCVC.boxplot(showmeans=True,ax=ax[0],grid=False)
ax[0].set_title('Frequency of consumption of vegetables\n')

plt.tight_layout()
```

### LOCAL_ANALYSIS/Bad_eating_habits

```python
CAEC_no = len(df.loc[df['CAEC'] == 'no'])
CAEC_Sometimes = len(df.loc[df['CAEC'] == 'Sometimes'])
CAEC_Frequently = len(df.loc[df['CAEC'] == 'Frequently'])
CAEC_Always = len(df.loc[df['CAEC'] == 'Always'])

FAVC_yes = len(df.loc[df['FAVC'] == 'yes'])
FAVC_no = len(df.loc[df['FAVC'] == 'no'])

fig, ax = plt.subplots(1,2, figsize=(15,10))

ax[0].pie([FAVC_yes,FAVC_no],
    labels=['Yes','No'], 
    autopct='%1.1f%%',explode=[0.2,0], shadow=True)
ax[0].set_title('Frequent consumption of high caloric food')

ax[1].bar(['no','Sometimes','Frequently','Always'],
    [CAEC_no,CAEC_Sometimes,CAEC_Frequently,CAEC_Always], 
    color = 'cyan')
ax[1].set_title('Consumption of food between meals\n')


ax[0].legend(loc = 3,fontsize = 12)

plt.tight_layout()
```

### LOCAL_ANALYSIS/Quantification_food_consomption

```python
SCC_yes = len(df.loc[df['SCC'] == 'yes'])
SCC_no = len(df.loc[df['SCC'] == 'no'])

df_NCP = df[["NCP"]]

fig, ax = plt.subplots(1,2, figsize=(15,10))

ax[0].pie([SCC_yes,SCC_no],
    labels=['Yes','No'],
    autopct='%1.1f%%',explode=[0.1,0],startangle = 30,shadow = True)
ax[0].set_title("Calories's consumption monitoring")

sns.histplot(df_NCP,color='cyan',ax = ax[1],kde=True)
ax[1].set_title("Number of main meals\n")

ax[0].legend(loc=3)
```

### LOCAL_ANALYSIS/Lifestyle_habits

```python
Automobile = len(df.loc[df['MTRANS'] == 'Automobile'])
Motorbike = len(df.loc[df['MTRANS'] == 'Motorbike'])
Bike = len(df.loc[df['MTRANS'] == 'Bike'])
Public_Transportation  = len(df.loc[df['MTRANS'] == 'Public_Transportation'])
Walking = len(df.loc[df['MTRANS'] == 'Walking'])

plt.bar(['Automobile','Motorbike','Bike','Public_Transp','Walking'],
    [Automobile,Motorbike,Bike,Public_Transportation,Walking],
    color = 'orange')
plt.title('Type of transportation used')
```

```python
df_FAF = df[["FAF"]]
df_TUE = df[["TUE"]]

fig, ax = plt.subplots(2,2, figsize=(15,10))

df_TUE.boxplot(showmeans=True,grid = False,ax=ax[0][0])
ax[0][0].set_title('Time using technology devices\n')

df_FAF.boxplot(showmeans=True,grid = False,ax=ax[1][0])
ax[1][0].set_title('\nPhysical activity frequency\n')

sns.histplot(df_TUE, ax= ax[0][1], kde=True)
ax[0][1].set_title('\nTime using technology devices\n')

sns.histplot(df_FAF, ax= ax[1][1], kde=True)
ax[1][1].set_title('\nPhysical activity frequency\n')

plt.tight_layout()
```

### GLOBAL_ANALYSISGLOBAL_ANALYSIS/Insufficient_Weight

```python
df_Insufficient_Weight=df[df["NObeyesdad"]=="Insufficient_Weight"]
df_Insufficient_Weight

#Condition where remplit les lignes ne respectant pas les conditions avec des NaN, on les supprime ensuite avec dropna(). 
df_Insufficient_Weight=df[["Gender","Age","Height","Weight","family_history_with_overweight"]].where(df["NObeyesdad"]=="Insufficient_Weight").dropna()
df_Insufficient_Weight_Female=df_Insufficient_Weight[["Age","Height","Weight","family_history_with_overweight"]].where(df_Insufficient_Weight["Gender"]=="Female").dropna()
df_Insufficient_Weight_Male=df_Insufficient_Weight[["Age","Height","Weight","family_history_with_overweight"]].where(df_Insufficient_Weight["Gender"]=="Male").dropna()

fg, ax =plt.subplots(3,4,figsize=(30,16))
df_Insufficient_Weight.boxplot(df_Insufficient_Weight.columns.tolist()[1],showmeans=True,ax=ax[0,0])
df_Insufficient_Weight.boxplot(df_Insufficient_Weight.columns.tolist()[2],showmeans=True,ax=ax[1,0])
df_Insufficient_Weight.boxplot(df_Insufficient_Weight.columns.tolist()[3],showmeans=True,ax=ax[2,0])
sns.countplot(x="Gender", hue="family_history_with_overweight", data=df_Insufficient_Weight,ax=ax[0,3])

df_Insufficient_Weight_Female.boxplot(df_Insufficient_Weight_Female.columns.tolist()[0],showmeans=True,ax=ax[0,1])
df_Insufficient_Weight_Female.boxplot(df_Insufficient_Weight_Female.columns.tolist()[1],showmeans=True,ax=ax[1,1])
df_Insufficient_Weight_Female.boxplot(df_Insufficient_Weight_Female.columns.tolist()[2],showmeans=True,ax=ax[2,1])
sns.countplot(x="family_history_with_overweight", data=df_Insufficient_Weight_Female,ax=ax[1,3])

df_Insufficient_Weight_Male.boxplot(df_Insufficient_Weight_Male.columns.tolist()[0],showmeans=True,ax=ax[0,2])
df_Insufficient_Weight_Male.boxplot(df_Insufficient_Weight_Male.columns.tolist()[1],showmeans=True,ax=ax[1,2])
df_Insufficient_Weight_Male.boxplot(df_Insufficient_Weight_Male.columns.tolist()[2],showmeans=True,ax=ax[2,2])
sns.countplot(x="family_history_with_overweight", data=df_Insufficient_Weight_Male,ax=ax[2,3])


ax[0,0].set_title('Female  and male Age with Insufficient_Weight')
ax[0,0].set_ylabel('Age in years  ')

ax[1,0].set_title('Female and male Height with Insufficient_Weight')
ax[1,0].set_ylabel('Height in cm ')

ax[2,0].set_title('Female and male Weight with Insufficient_Weight')
ax[2,0].set_ylabel('Weight in Kg')
ax[0,3].set_title('Countplot of female and male  with Insufficient_Weight in fonction of their family history with alcohol')


ax[0,1].set_title('Female Age with Insufficient_Weight')
ax[0,1].set_ylabel('Age in years  ')

ax[1,1].set_title('Female Height with Insufficient_Weight')
ax[1,1].set_ylabel('Height in cm ')

ax[2,1].set_title('Female Weight with Insufficient_Weight')
ax[2,1].set_ylabel('Weight in Kg')
ax[1,3].set_title('Countplot of female  with Insufficient_Weight in fonction of their family history with alcohol')


ax[0,2].set_title('Male Age with Insufficient_Weight')
ax[0,2].set_ylabel('Age in years  ')

ax[1,2].set_title('Male Height with Insufficient_Weight')
ax[1,2].set_ylabel('Height in cm ')

ax[2,2].set_title('Male Weight with Insufficient_Weight')
ax[2,2].set_ylabel('Weight in Kg')
ax[2,3].set_title('Countplot of male  with Insufficient_Weight in fonction of their family history with alcohol')

```
